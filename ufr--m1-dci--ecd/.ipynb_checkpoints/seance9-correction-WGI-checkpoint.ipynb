# Séance 9 -- correction -- Données WGI

```python
import seaborn
import pandas
from sklearn import manifold
import seaborn.objects as so
```

## Données WGI


```python
wgi = pandas.read_csv("https://fxjollois.github.io/donnees/WGI/wgi2019.csv")
wgi.head(10)
```

```python
wgi_complet = wgi[~wgi.isna().any(axis = 1)].reset_index()
wgi_num = wgi_complet[wgi.columns[2:]]
wgi_num
```

### ISOMAP

```python
isomap = manifold.Isomap(n_components = 2, n_neighbors = 25)
res = isomap.fit_transform(wgi_num)

so.Plot(pandas.DataFrame(res, columns = ["X", "Y"]).assign(Pays = wgi_complet["Country"]), 
        x = "X", y = "Y", text = "Pays") \
    .add(so.Text()) \
    .theme({"axes.facecolor": "w", "axes.edgecolor": "slategray"}) \
    .layout(size=(16, 16))
```

### LLE

```python
lle = manifold.LocallyLinearEmbedding(n_components = 2, n_neighbors = 25)
res = lle.fit_transform(wgi_num)

so.Plot(pandas.DataFrame(res, columns = ["X", "Y"]).assign(Pays = wgi_complet["Country"]), 
        x = "X", y = "Y", text = "Pays") \
    .add(so.Text()) \
    .theme({"axes.facecolor": "w", "axes.edgecolor": "slategray"}) \
    .layout(size=(16, 16))
```

### MDS

```python
mds = manifold.MDS(n_components = 2, normalized_stress = "auto", random_state = 123456)
res = mds.fit_transform(wgi_num)

so.Plot(pandas.DataFrame(res, columns = ["X", "Y"]).assign(Pays = wgi_complet["Country"]), 
        x = "X", y = "Y", text = "Pays") \
    .add(so.Text()) \
    .theme({"axes.facecolor": "w", "axes.edgecolor": "slategray"}) \
    .layout(size=(16, 16))
```
