# Séance 9 -- correction -- Données Wine

```python
import seaborn
import pandas
from sklearn import manifold
import seaborn.objects as so
from sklearn.preprocessing import scale
```

## Données Wine


```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
wine = pandas.read_csv(url, header = None, sep = ",")
wine.columns = ["class", "Alcohol", "Malic acid", "Ash", "Alcalinity of ash", "Magnesium", 
                "Total phenols", "Flavanoids", "Nonflavanoid phenols", "Proanthocyanins", 
                "Color intensity", "Hue", "OD280/OD315 of diluted wines", "Proline"]
wine
```

```python
wine_bis = scale(wine.drop(columns = "class"))
wine_bis
```

### ISOMAP

```python
isomap = manifold.Isomap(n_components = 2, n_neighbors = 25)
res = isomap.fit_transform(wine_bis)

res_df = pandas.DataFrame(res, columns = ["X", "Y"])
res_df["Classe"] = [str(e) for e in wine["class"]]
so.Plot(res_df, 
        x = "X", y = "Y", color = "Classe") \
    .add(so.Dot()) \
    .theme({"axes.facecolor": "w", "axes.edgecolor": "slategray"}) \
    .layout(size=(12, 6))
```

### LLE

```python
lle = manifold.LocallyLinearEmbedding(n_components = 2, n_neighbors = 20)
res = lle.fit_transform(wine_bis)

res_df = pandas.DataFrame(res, columns = ["X", "Y"])
res_df["Classe"] = [str(e) for e in wine["class"]]
so.Plot(res_df, 
        x = "X", y = "Y", color = "Classe") \
    .add(so.Dot()) \
    .theme({"axes.facecolor": "w", "axes.edgecolor": "slategray"}) \
    .layout(size=(12, 6))
```

### MDS

```python
mds = manifold.MDS(n_components = 2, normalized_stress = "auto", random_state = 123456)
res = mds.fit_transform(wine_bis)

res_df = pandas.DataFrame(res, columns = ["X", "Y"])
res_df["Classe"] = [str(e) for e in wine["class"]]
so.Plot(res_df, 
        x = "X", y = "Y", color = "Classe") \
    .add(so.Dot()) \
    .theme({"axes.facecolor": "w", "axes.edgecolor": "slategray"}) \
    .layout(size=(12, 6))
```
