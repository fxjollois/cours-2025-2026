```python
import pandas

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/ozone/onehr.data"
data = pandas.read_csv(url, header = None, na_values = "?")
data
```

## Chargement des différents modules utiles

```python
import numpy
import matplotlib.pyplot as plt
import seaborn
seaborn.set_style("white")

from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from sklearn.metrics import roc_curve, auc
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.neural_network import MLPClassifier
```

## Création de deux objets `X` et `y` ppour simplification du code par la suite

```python
X = data.drop(["Date", "ozone"], axis = 1)
X.head()
```

```python
y = data["ozone"]
y
```

## 3- ACP

```python
pca = PCA().fit(scale(X))
```

```python
eig = pandas.DataFrame(
    {
        "Dimension" : ["Dim" + str(x + 1) for x in range(X.shape[1])], 
        "Valeur propre" : pca.explained_variance_,
        "% variance expliquée" : numpy.round(pca.explained_variance_ratio_ * 100),
        "% cum. var. expliquée" : numpy.round(numpy.cumsum(pca.explained_variance_ratio_) * 100)
    },
    columns = ["Dimension", "Valeur propre", "% variance expliquée", "% cum. var. expliquée"]
)
eig
```

```python
plt.figure(figsize=(16, 6))
g_eig = seaborn.barplot(x = "Dimension", 
                        y = "% variance expliquée",
                        color = "lightseagreen",
                        data = eig)
plt.text(12, 100/X.shape[1] + 1, "Seuil")
plt.axhline(y = 100/X.shape[1], linewidth = .5, color = "dimgray", linestyle = "--")
g_eig.set(ylabel = "Variance expliquée (%)")
g_eig.figure.suptitle("Variance expliquée par dimension")

plt.show()
```

Dans le tableau et le graphique ci-dessous, nous remarquons que les 2 premières composantes permettent d'expliquer à elles-seules **63% de l'information**.

Par contre, dans le plan factoriel ci-dessous, nous ne remarquons pas de structure en classe.

```python
X_pca_row = pandas.DataFrame(pca.transform(scale(X)), 
                             columns = ["Dim" + str(i+1) for i in range(X.shape[1])])
g_pca = seaborn.lmplot(data = X_pca_row, x = "Dim1", y = "Dim2", height = 6, aspect = 2, fit_reg = False)
plt.show()
```

Si nous regardons le cercle des corrélations, nous remarquons que certaines variables sont très proches entre elles. Cela s'explique par le fait qu'il y ait beaucoup de variables, avec certains très corrélées entre elles (cf graphique ci-après - en particulier les variables `Tx`, mais aussi les `WSRx`). Dans un travail complet, nous devrions sélectionner uniquement certaines variables (par exemple `WSR_PK` et `T_PK`).

```python
coordvar_df = pandas.DataFrame(pca.components_.T * numpy.sqrt(pca.explained_variance_), 
                               columns=['PC'+str(i+1) for i in range(X_pca_row.shape[1])], 
                               index=X.columns)

fig, axes = plt.subplots(figsize = (10, 10))
fig.suptitle("Cercle des corrélations")
axes.set_xlim(-1, 1)
axes.set_ylim(-1, 1)
axes.axvline(x = 0, color = 'lightgray', linestyle = '--', linewidth = 1)
axes.axhline(y = 0, color = 'lightgray', linestyle = '--', linewidth = 1)
for j in range(X.shape[1]):
    axes.text(coordvar_df["PC1"].iloc[j],coordvar_df["PC2"].iloc[j], coordvar_df.index[j], size = 25)
    axes.plot([0,coordvar_df["PC1"].iloc[j]], [0,coordvar_df["PC2"].iloc[j]], color = "gray", linestyle = 'dashed')
plt.gca().add_artist(plt.Circle((0,0),1,color='blue',fill=False))

plt.show()
```

```python
plt.figure(figsize=(8, 8))
seaborn.heatmap(X.corr(), vmin = -1, vmax = 1, cmap = "Spectral")
plt.show()
```

## 4- Variable cible sur ACP

## 5- Partition

```python
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters = k, init = "random", n_init = 20).fit(scale(X))
    inertia = inertia + [kmeans.inertia_]
    
rsquare = [(inertia[0] - i) / inertia[0] for i in inertia]
criteres = pandas.DataFrame({
    "k": range(1, 11), 
    "inertia": inertia,
    "rsquare": rsquare,
    "pseudof": [(rsquare[k-1] / k) / ((1 - rsquare[k-1]) / (X.shape[0] - k)) if k > 1 else None for k in range(1, 11)]
})

g = seaborn.FacetGrid(data = criteres.melt(id_vars = "k"), col = "variable", sharey = False, 
                      height = 5, aspect = 1)
g.map_dataframe(seaborn.lineplot, x = "k", y = "value")
g.map(plt.axvline, x = 2, linewidth = .5, color = "dimgray", linestyle = "--")
g.map(plt.axvline, x = 3, linewidth = .5, color = "dimgray", linestyle = "--")
g.add_legend()
plt.show()
```

A l'aide de $k$-means, et des graphiques ci-desssus, on peut choisir soit 2 classes (en se basant sur l'évolution de l'inertie et du $R^2$) ou 3 classes (en se basant sur le $PseudoF$). Nous allons choisir ici **3 classes**.

```python
kmeans = KMeans(n_clusters = 3, init = "random", n_init = 20).fit(scale(X))
```

Les classes sont clairement séparées sur le plan factoriel. La plupart des jours avec dépassement d'ozone sont principalement dans une même classe.

```python
g_pca = seaborn.lmplot(
    data = X_pca_row.assign(classe = kmeans.labels_),
    x = "Dim1", y = "Dim2", hue = "classe", fit_reg = False, 
    height = 6, aspect = 2)
plt.show()
```

```python
pandas.crosstab(kmeans.labels_, y)
```

## 6- Modélisation avec les 4 algos

### Régression logistique

```python
m_log = LogisticRegression(max_iter = 10000).fit(X, y)
```

```python
pandas.crosstab(m_log.predict(X), y)
```

```python
m_log.score(X, y)
```

### $k$-plus proches voisins

```python
r2 = []

k_val = (1, 2, 3, 4, 5, 10, 15, 20)
for k in k_val:
    m_knn = KNeighborsClassifier(n_neighbors = k).fit(scale(X), y)
    r2 = r2 + [m_knn.score(scale(X), y)]
    
df = pandas.DataFrame({
    "variable": k_val,
    "r2": r2
})
df
```

```python
m_knn = KNeighborsClassifier(n_neighbors = 3).fit(scale(X), y)
```

```python
pandas.crosstab(m_knn.predict(scale(X)), y)
```

```python
m_knn.score(scale(X), y)
```

### CART

```python
m_tree = DecisionTreeClassifier().fit(X, y)
```

```python
pandas.crosstab(m_tree.predict(X), y)
```

```python
m_tree.score(X, y)
```

```python
pruning = m_tree.cost_complexity_pruning_path(X, y)

fig, ax = plt.subplots()
ax.plot(pruning.ccp_alphas[:-1], pruning.impurities[:-1], marker="o")
ax.set_xlabel("effective alpha")
ax.set_ylabel("total impurity of leaves")
ax.set_title("Total Impurity vs effective alpha for training set")
plt.axvline(x = 0.00125, linewidth = .5, color = "dimgray", linestyle = "--")
plt.show()
```

Avec le graphique ci-dessus, on propose de couper à $cp = 0.00125$ pour élaguer l'arbre complet.

```python
m_tree_elague = DecisionTreeClassifier(ccp_alpha=0.00125).fit(X, y)
```

```python
pandas.crosstab(m_tree_elague.predict(X), y)
```

```python
m_tree_elague.score(X, y)
```

```python
plt.figure(figsize=(16, 16))
plot_tree(m_tree_elague, feature_names = X.columns)
plt.show()
```

### MLP

```python
r2 = []
auc_list = []

layers = ((), (10), (20), (50), (100), (10, 10), (20, 20), (50, 50), (100, 100), 
          (10, 10, 10), (20, 20, 20), (50, 50, 50), (100, 100, 100))
for l in layers:
    m_mlp = MLPClassifier(hidden_layer_sizes = l, max_iter = 1000).fit(X, y)
    r2 = r2 + [m_mlp.score(X, y)]
    fpr, tpr, th = roc_curve(y, pandas.DataFrame(m_mlp.predict_proba(X))[1])
    auc_list = auc_list + [auc(fpr, tpr)]
    
df = pandas.DataFrame({
    "layers": layers,
    "r2": r2,
    "auc": auc_list
})
df
```

```python
m_mlp = MLPClassifier(hidden_layer_sizes = (10, 10), max_iter = 1000).fit(X, y)
```

```python
pandas.crosstab(m_mlp.predict(X), y)
```

```python
m_mlp.score(X, y)
```

D'après les résultats ci-dessus, nous remarquons les éléments suivants :

- Le perceptron multi-couches obtenu fait énormément d'erreurs et n'est pas intéressant ici ;
- La régression logistique fonctionne assez bien, avec toutefois beaucoup de faux négatifs ;
- La méthode des $k$-plus proches voisins fonctionne un peu mieux, avec toujours beaucoup de faux négatifs ;
- L'arbre élagué permet de faire moins d'erreurs.
